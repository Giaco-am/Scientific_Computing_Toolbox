```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import interpolation_bindings as ib 
```

```python
quadratic_data = pd.read_csv("../data/quadratic_data.csv")
quadratic_points = ib.df_to_set_of_points(quadratic_data)

cubic_data = pd.read_csv("../data/cubic_data.csv")
cubic_points = ib.df_to_set_of_points(cubic_data)
```

```python
plt.plot(quadratic_data["x"], quadratic_data["y"], 'ro', label="Quadratic Data")
```

```python
plt.plot(cubic_data["x"], cubic_data["y"], 'bo', label="Cubic Data")
```

```python
linear_interpolation = ib.LinearInterpolation(quadratic_points)
```

```python
## creating 1000 values between the min and max of the x values of the data
x_values = np.linspace(quadratic_data["x"].min(), quadratic_data["x"].max(), 1000)
y_values = [linear_interpolation.interpolate(x) for x in x_values]
plt.plot(x_values, y_values, 'r-', label="Linear Interpolation")
## plotting the initial points
plt.plot(quadratic_data["x"], quadratic_data["y"], 'ro')
```

```python
## creating 10 random values with mean 0
y = np.random.randn(10)
## taking as x the values from 1 to 10
x = np.arange(1, 11)
random_data = pd.DataFrame({"x": x, "y": y})
random_points = ib.df_to_set_of_points(random_data)
```

```python
plt.plot(random_data["x"], random_data["y"], 'go', label="Random Data")
```

```python
lagrange_interpolation = ib.Lagrange(random_points)
```

```python
## creating 1000 values between the min and max of the x values of the data
x_values = np.linspace(quadratic_data["x"].min(), quadratic_data["x"].max(), 1000)
y_values = [lagrange_interpolation.interpolate(x) for x in x_values]
plt.plot(x_values, y_values, 'r-', label="Lagrange Interpolation")
## plotting the initial points
plt.plot(random_data["x"], random_data["y"], 'go')
```
